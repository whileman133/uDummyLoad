```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import ScalarFormatter
from itertools import cycle

ls = cycle(['--', '-', '-.'])
col = cycle(['blue', 'black', 'red'])

Cb = 4.3 # J/C
reb = 3.0 # C/W
Temax = 110 # C
TTA = [0.0, 25.0, 40.0] # C
PP = np.arange(5, 20, 0.01) # W

fig, ax = plt.subplots(dpi=300, figsize=(5, 5))

for TA in TTA:
    ttmax = Cb * ( (Temax - TA) / PP - reb)
    ax.semilogy(PP, ttmax, label=f"$T_A={TA:.1f}$", ls=next(ls), color=next(col))
    
ax.yaxis.set_major_formatter(ScalarFormatter())
plt.ylim([1, 100])
plt.xlim([5, 20])
plt.xticks(np.arange(min(PP), max(PP)+1, 1.0))
plt.legend()
plt.grid(which='both')
plt.xlabel('RF Input Power $P_{in}$ (Watts)')
plt.ylabel('Safe Operating Time (seconds)')
fig.tight_layout()
plt.savefig('safe-operating-temp.png')
```
